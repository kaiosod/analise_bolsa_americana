# Vizualização de Dados

```python
import pandas as pd
import matplotlib.pyplot as plt 

base = pd.read_csv('base.csv')

base
```

```python
# Plotar uma lista com as ações que mais valorizaram 

base_mais_valorizados = base.sort_values(by='Appreciation',ascending=False).head(10)

base_mais_valorizados
```

```python
# As 10 empresas que mais valorizaram

base_mais_valorizados.plot.bar('Name','Appreciation',color='royalblue',ec='k',alpha =.7)
plt.xlabel('Empresas')
plt.ylabel('Valorização')
plt.title('As 10 Empresas que mais valorizaram')
plt.savefig("top10.png")

#plt.style.use("ggplot")
```

```python
# Preço de Mercado de cada setor

base_setores = base.groupby('Sector').count().reset_index()

base_setores.plot.bar('Sector','Market Cap',color='royalblue',ec='k',alpha =.7)
plt.xlabel('Setores')
plt.ylabel('Valor de mercado em Dólares')
plt.savefig("setor.png")
```

```python
# As 15 empresas com maior valor de mercado 

base_maiores_mercado = base.sort_values(by='Market Cap',ascending=False).head(15)

base_maiores_mercado.plot.barh('Name','Market Cap',color='royalblue',ec='k',alpha =.7)
plt.title('As 15 empresas com maior valor de mercado')
plt.xlabel('Valor de Mercado em Dólares')
plt.ylabel('Empresas')
plt.savefig("maiores15.png")
```

```python
# As 3 maiores empresas de cada setor

base2 = base.sort_values(by='Market Cap',ascending=False)

base_industria = base2[base2['Sector'] == 'Industrials']
base_financeiro = base2[base2['Sector'] == 'Financials' ]
base_consumo = base2[base2['Sector'] == 'Consumer Discretionary']
base_staple = base2[base2['Sector'] == 'Consumer Staples']
base_energia = base2[base2['Sector'] == 'Energy']
base_saude = base2[base2['Sector'] == 'Health Care']
base_ti = base2[base2['Sector'] == 'Information Technology']
base_material = base2[base2['Sector'] == 'Materials']
base_estado = base2[base2['Sector'] == 'Real Estate']
base_tele = base2[base2['Sector'] == 'Telecommunication Services']
base_utilidades = base2[base2['Sector'] == 'Utilities']

base_industria = base_industria.sort_values(by='Market Cap', ascending=False).head(1)
base_financeiro = base_financeiro.sort_values(by='Market Cap', ascending=False).head(1)
base_consumo = base_consumo.sort_values(by='Market Cap', ascending=False).head(1)
base_staple = base_staple.sort_values(by='Market Cap', ascending=False).head(1)
base_energia = base_energia.sort_values(by='Market Cap', ascending=False).head(1)
base_saude = base_saude.sort_values(by='Market Cap', ascending=False).head(1)
base_ti = base_ti.sort_values(by='Market Cap', ascending=False).head(1)
base_material = base_material.sort_values(by='Market Cap', ascending=False).head(1)
base_estado = base_estado.sort_values(by='Market Cap', ascending=False).head(1)
base_tele = base_tele.sort_values(by='Market Cap', ascending=False).head(1)
base_utilidades = base_utilidades.sort_values(by='Market Cap', ascending=False).head(1)

base_nova = pd.concat([base_industria,base_financeiro,base_consumo,base_staple,base_energia,base_saude,base_ti,base_material,
                      base_estado,base_tele,base_utilidades])

base_nova['Empresa'] = base_nova['Name'] +" / "+base_nova['Sector']

base_nova
```

```python
base_nova.plot.barh('Empresa','Market Cap',color='royalblue',ec='k',alpha =.7)
plt.xlabel('Valor de Mercado')
plt.ylabel('Empresas')
plt.title('Maiores Empresas de cada setor')
plt.savefig("maiores_setor.png")
```
